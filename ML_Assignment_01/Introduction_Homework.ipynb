## Task 4:- Gaussian Distribution[40 points]
#### Your final output should look like below 


#### Consider a Gaussian distribution $p_x(x)$ over x with mean  = 6 and standard deviation $\sigma$ = 1. Plot this distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

def pdf(x, mean, std):
    y_out = 1/(std * np.sqrt(2 * np.pi)) * np.exp( - (x - mean)**2 / (2 * std**2))
    return y_out

mean = 6.0
std = 1.0

indexes = np.arange(0, 12, 0.001)
bell_curve = pdf(x, mean, std)

fig, ax = plt.subplots(figsize=(16,9))
ax.plot(indexes, bell_curve)
ax.set_title("Bell curve with mean=6 and std=1")
ax.set_ylabel("Frequency")
ax.set_ylabel("Range")
ax.grid()

plt.show()
```

#### Draw a sample of N = 50000 points from this distribution and plot a histogram of their values, which as expected agrees with the distribution $p_x(x)$. 

```python
data = np.random.randn(50000) * std + mean

fig, ax = plt.subplots(figsize=(9,6))

ax.plot(indexes,bell_curve)
ax.hist(data, bins=21, normed=1,color="silver")

ax.set_title("Overlay histogram and bell curve")
ax.set_ylabel("Frequency")
ax.set_xlabel("Range")
ax.grid()

plt.show()
```

#### Now consider a non-linear change of variables from x to y given by  
#### $x=g(y)=ln(y)-ln(1-y)+5$  
#### The inverse of this function is given by  
#### $y=g^{-1}(x)=\frac{1}{1+exp(-x+5)}$  
#### which is a logistic sigmoid function. Plot this function along with the previous plot

```python
def logistic_sigmoid(x):
    return np.log(x) - np.log(1-x) + 5

def inverse_logistic_sigmoid(x):
    return 1.0 / (1 + np.exp(-x + 5))

log = logistic_sigmoid(bell_curve)
inv = inverse_logistic_sigmoid(indexes)

plt.plot(bell_curve, bell_curve * logistic_sigmoid(log))
```
