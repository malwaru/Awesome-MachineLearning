## Task 1:- Find the box[20 points]

#### Suppose that we have three colored boxes r (red), b (blue), and g (green).
#### - Box r contains 3 apples, 4 oranges, and 3 limes, 
#### - Box b contains 1 apple, 1 orange, and 0 limes, and 
#### - Box g contains 3 apples, 3 oranges, and 4 limes. 
#### soln1: If a box is chosen at random with probabilities p(r) = 0.2, p(b) = 0.2, p(g) = 0.6, and a piece of fruit is removed from the box (with equal probability of selecting any of the items in the box), then what is the probability of selecting an apple? 
#### soln2: If we observe that the selected fruit is, in fact an orange, what is the probability that it came from the green box?

**Give your mathematical writing here**



Let probability of choosing an
apple   is P(a)

$T_b$=Total no of fruits in box b

\begin{equation}

##### Solution 2

P(a)=P(r)*\frac{a}{T_r}+P(b)*\frac{a}{T_b}+P(g)*\frac{a}{T_g}

## Task 3:- Gmail spam filter[20 points]

#### Gmail is used widely across the globe and the amount of Machine Learning incorporation is increasing periodically. One of the very useful incorporations is ham (non-spam)/spam classification. From past experience, we estimated that 50% of the emails in the inbox are spam. Gmail uses a Bayes classifier that filters out such spam before reaching your inbox. It has claimed that Gmail can detect 99% of spam emails, which accounts for 5% of false positive. If the upcoming mail is detected as spam, what is the probability that it is a ham?



**Give your mathematical writing here**

let $\textbf{z}$ be observed/detected result and $\textbf{x}$ be the actual result. $\textbf{S}$ represent spam and $\textbf{H}$ represent Ham

using the data provided writing down the given and inferred probabilities
$$P(S)=0.5 \rightarrow P(H)=0.5$$
$$P(z=S|x=H)=0.05 \rightarrow P(z=H|x=H)=0.95 $$
$$P(z=S|x=S)=0.99 \rightarrow P(z=H|x=S)=0.01 $$

Since we are supposed to find probability of a mail being ham given its detected as a spam

$$P(x=H|z=S)=\frac{P(z=S|x=H)P(x=H)}{P(z=S)}$$
$$$$
$$P(x=H|z=S)=\frac{P(z=S|x=H)P(x=H)}{P(z=S|x=S)P(S)+P(z=S|x=H)P(H)}$$

```python
##Storing data defined above as a matrix
#Prior priors in order of[P(S), P(H)]

P=[0.5,0.5]

#Detector model in order[[P(z=S|x=H),P(z=H|x=H)];[P(z=S|x=S),P(z=H|x=S)]]
P_Z=[[0.05,0.95],[0.99,0.01]]



soln4 =P_Z[0][0]*P[1]/((P_Z[0][0]*P[0])+(P_Z[1][0]*P[1]))
print(soln4)
```

## Task 4:- Gaussian Distribution[40 points]
#### Your final output should look like below 


#### Consider a Gaussian distribution $p_x(x)$ over x with mean  = 6 and standard deviation $\sigma$ = 1. Plot this distribution.

```python
mue=6
sigma=1

x=np.linspace(0,10,50000)
y=(2*x)+8
def gaussian(x,mue,sigma):
    return (1/(sigma*(2*np.pi)**0.5))*np.exp(-0.5*((x-mue)/sigma)**2)



P_X=gaussian(x,mue,sigma)

figure1=plt.figure(figsize=(15,9))
ax1=figure1.add_subplot(111)

ax1.plot(x,P_X,label='p(x)')
ax1.set(title="Gaussian Probability",xlabel='Data')
ax1.legend()
ax1.grid()

plt.show()
```

#### Draw a sample of N = 50000 points from this distribution and plot a histogram of their values, which as expected agrees with the distribution $p_x(x)$. 

```python
N_points = 50000
n_bins = 30

# Generate a normal distribution, center at x=6
x1 = np.random.randn(N_points)+6




## We the density true function normalize the PDF
figure2=plt.figure(figsize=(15,9))
ax2=figure2.add_subplot(111)
ax2.plot(x,P_X,label='p(x)')
ax2.hist(x1, bins=n_bins,density=True,ec='black',label='Histogram')
ax2.set(title="Gaussian Probability",xlabel='Data')
ax2.legend()
ax2.grid()
plt.show()


```

#### Now consider a non-linear change of variables from x to y given by  
#### $x=g(y)=ln(y)-ln(1-y)+5$  
#### The inverse of this function is given by  
#### $y=g^{-1}(x)=\frac{1}{1+exp(-x+5)}$  
#### which is a logistic sigmoid function. Plot this function along with the previous plot

```python
y=1/(1+np.exp(-x+5))

figure3=plt.figure(figsize=(15,9))
ax3=figure3.add_subplot(111)
ax3.plot(x,P_X,label='p(x)')
ax3.hist(x1, bins=n_bins,density=True,ec='black',label='Histogram')
ax3.plot(x,y,label='y(x)')
ax3.set(title="Gaussian Probability",xlabel='Data')
ax3.legend()
ax3.grid()
plt.show()
```

#### Plot the curve $p_x(g(y))$ that is a transformation of $p_x(x)$ as a function of x, $p_x(g(y))$ along with the previous plots, this should look exactly as the above one(reference provided above).

```python
g_y=np.log(y)-np.log(1-y)+5
p_gy=g_y*P_X



figure4=plt.figure(figsize=(15,9))
ax4=figure4.add_subplot(111)
ax4.plot(x,P_X,label='p(x)')
ax4.hist(x1, bins=n_bins,density=True,ec='black',label='Histogram')
ax4.plot(x,y,label='y(x)')


ax4.plot(x,p_gy,'b--',label='p(g(y))')


ax4.set(title="Gaussian Probability",xlabel='Data')
ax4.legend()
ax4.grid()
plt.show()







```
