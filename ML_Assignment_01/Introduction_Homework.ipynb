## Task 1:- Find the box[20 points]

#### Suppose that we have three colored boxes r (red), b (blue), and g (green).
#### - Box r contains 3 apples, 4 oranges, and 3 limes, 
#### - Box b contains 1 apple, 1 orange, and 0 limes, and

#### soln1: If a box is chosen at random with probabilities p(r) = 0.2, p(b) = 0.2, p(g) = 0.6, and a piece of fruit is removed from the box (with equal probability of selecting any of the items in the box), then what is the probability of selecting an apple? 
#### soln2: If we observe that the selected fruit is, in fact an orange, what is the probability that it came from the green box?

Box R:
+  3 Apples $\rightarrow P(A|R) = 0.3$
+  4 Oranges $\rightarrow P(O|R) = 0.4$
+  3 Limes   $\rightarrow P(L|R) = 0.3$
+  10 Total
+  $P(R) = 0.2$

Box B:
+  1 Apples $\rightarrow P(A|B) = 0.5$
+  1 Oranges $\rightarrow P(0|B) = 0.5$
+  0 Limes $\rightarrow P(L|B) = 0$
+  2 Total
+  $P(B) = 0.2$

Box G:
+  3 Apples $\rightarrow P(A|G) = 0.3$
+  3 Oranges $\rightarrow P(O|G) = 0.3$
+  4 Limes $\rightarrow P(L|G) = 0.4$
+  10 Total
+  $P(G) = 0.6$

First solution: 

$P(A) = P(A|R) * P(R) + P(A|B) * P(B) + P(A|G) * P(B) \rightarrow P(A) = 0.3 * 0.2 + 0.5 * 0.2 + 0.3 * 0.6$

----

----

## Task 4:- Gaussian Distribution[40 points]
#### Your final output should look like below 


#### Consider a Gaussian distribution $p_x(x)$ over x with mean μ = 6 and standard deviation $\sigma$ = 1. Plot this distribution.

```python
def gaussian(x,mue,sigma):
    '''
    The function to output a gaussian distribution
    
    Parameters
    ----------
    x    : The input data point ->float array
    mue  : The mean             ->float 
    sigma: The variance         ->float
    
    
    Returns
    ----------
    The gaussian output         ->float array
    
    
    '''
    return (1/(sigma*(2*np.pi)**0.5))*np.exp(-0.5*((x-mue)/sigma)**2)

##Defining the mean and variance

mue=6
sigma=1


## Creating the x variables
x=np.linspace(0,10,50000)




##Create P_x gaussian 
P_X=gaussian(x,mue,sigma)

#Plotting Graph

figure1=plt.figure(figsize=(15,9))
ax1=figure1.add_subplot(111)
ax1.plot(x,P_X,label='p(x)')
ax1.set(title="Gaussian Probability",xlabel='Data',ylabel='P(x)')
ax1.legend()
ax1.grid()
plt.show()
```

#### Draw a sample of N = 50000 points from this distribution and plot a histogram of their values, which as expected agrees with the distribution $p_x(x)$. 

```python
N_points = 50000
n_bins = 50

# Generate a normal distribution, center at x=6
x1 = np.random.randn(N_points)+6




## We use the density=true function normalize the histogram or else it will just be the counts
figure2=plt.figure(figsize=(15,9))
ax2=figure2.add_subplot(111)
ax2.plot(x,P_X,label='p(x)')
ax2.hist(x1, bins=n_bins,density=True,ec='black',label='Histogram')
ax2.set(title="Gaussian Probability along histogram",xlabel='Data',ylabel='P(x)')
ax2.legend()
ax2.grid()
plt.show()


```

#### Now consider a non-linear change of variables from x to y given by

#### The inverse of this function is given by  
#### $y=g^{-1}(x)=\frac{1}{1+exp(-x+5)}$  
#### which is a logistic sigmoid function. Plot this function along with the previous plot

```python
## Defining the inverse function g^1(x)
y=1/(1+np.exp(-x+5))



##Plotting all three graphs
figure3=plt.figure(figsize=(15,9))
ax3=figure3.add_subplot(111)
ax3.plot(x,P_X,label='p(x)')
ax3.hist(x1, bins=n_bins,density=True,edgecolor='black',label='Histogram')
ax3.plot(x,y,label='y(x)')
ax3.set(title="Gaussian Probability with histogram and logistic sigmoid",xlabel='Data',ylabel='P(x)')
ax3.legend()
ax3.grid()
plt.show()
```
